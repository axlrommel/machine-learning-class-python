```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
n = 40
x = np.linspace(0,10,n) + np.random.randn(n)/3
y = np.sin(x)+x/3 + np.random.randn(n)/3
data = np.array([x,y])
df = pd.DataFrame(data).T
df.to_csv('poly.txt')
plt.plot(x,y,'ro')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
data = pd.read_csv('ex1data2.txt')
fig = plt.figure()
X, Y = np.meshgrid(data['population'],data['sales'])
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,data['valy'], cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.show()
```
